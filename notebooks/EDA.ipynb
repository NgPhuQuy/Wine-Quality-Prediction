OUTLIERS

```python
Q1 = red_wine_df['alcohol'].quantile(0.25)
Q3 = red_wine_df['alcohol'].quantile(0.75)
IQR = Q3 - Q1

outliers = red_wine_df[(red_wine_df['alcohol'] < Q1 - 1.5*IQR) |
                       (red_wine_df['alcohol'] > Q3 + 1.5*IQR)]

print(outliers)
```
