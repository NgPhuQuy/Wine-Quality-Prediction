OUTLIERS

```python
Q1 = df_red_wine['alcohol'].quantile(0.25)
Q3 = df_red_wine['alcohol'].quantile(0.75)
IQR = Q3 - Q1

outliers = df_red_wine[(df_red_wine['alcohol'] < Q1 - 1.5*IQR) |
                       (df_red_wine['alcohol'] > Q3 + 1.5*IQR)]

print(outliers)
```
